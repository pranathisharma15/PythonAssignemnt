```python
def count_log_level(logs): 
    log_levels = {'INFO': 0, 'WARN': 0, 'ERROR': 0} 
    for log in logs: 
        level = log.split(":")[0]
        if level in log_levels: 
            log_levels[level] += 1 
    return log_levels

logs = {"ERROR: DISK FULL", "INFO: STARTED", "ERROR: TIMEOUT", "WARN: LOW MEMORY"}
print(count_log_level(logs))
```
