```python
def get_sla_hours(priority):
    sla_hours = {
        'P1': 4,
        'P2': 8,
        'P3': 24,
        'P4': 48
    }
    return -1 if priority not in sla_hours else sla_hours[priority]

print(get_sla_hours('P1')) 
print(get_sla_hours('P2')) 
print(get_sla_hours('P3')) 
print(get_sla_hours('P4')) 
```
